```python
import re
import pandas as pd
from pandas import read_excel

# skillsTable is the dataset containing skills. It has two columns:
# PreferredLabel contains the main label of each skill
# AlternativeLabel contains all the labels of each skill
def skill_string_extract(sentence, skillsTable):
  # sentence converted to lower case
  sentence = sentence.lower()
  # declaration of the list of skills extracted from the sentence
  extracted_skills = []
  # for each row of the skill dataset
  for i in range(0,len(skillsTable["AlternativeLabel"])):
    # both the alternative and the preferred label are saved
    altLabel = skillsTable["AlternativeLabel"][i]
    prefLabel = skillsTable["PreferredLabel"][i]
    # the pattern to be searched is the alternative label converted to lower case
    regex = altLabel.lower()
    # if the pattern is very short it is easy to match it inside words
    # so it is transformed into a regex with word boundaries before and after (\b)
    if len(regex) < 5:
      regex = r"\b" + regex + r"\b"
    # character "+" creates problems so it is deleted
    if "+" in regex:
      regex = regex.replace("+", "")
    # the extraction is performed
    extract = re.findall(regex, sentence)
    # if a match is found, the preferred label corresponding to the matched skill
    # is inserted in the extracted_skills list
    if extract != []:
      extracted_skills.append(prefLabel)
    # the list of extracted skills is returned
  return extracted_skills
```
